#### **Reading and checking data**


```python
import pandas as pd
```

```python
# Diretório onde os arquivos de dados do trabalho estão armazenados
data_dir = "./dados/"

arquivos = {
    2015: "contratos2015.xlsx",
    2016: "contratos2016.xlsx",
    2017: "contratos2017.xlsx",
}
# Criar DataFrames para cada ano
dfs = {}
for ano, arquivo in arquivos.items():
    dfs[ano] = pd.read_excel(data_dir + arquivo)

df_2015 = dfs[2015]
df_2016 = dfs[2016]
df_2017 = dfs[2017]
```

```python
arquivos = {
    2018: "contratos2018.xlsx",
    2019: "contratos2019.xlsx",
    2020: "contratos2020.xlsx",
}
# Criar DataFrames para cada ano
dfs = {}
for ano, arquivo in arquivos.items():
    dfs[ano] = pd.read_excel(data_dir + arquivo)

df_2018 = dfs[2018]
df_2019 = dfs[2019]
df_2020 = dfs[2020]
```

```python
arquivos = {
    2021: "contratos2021.xlsx",
    2022: "contratos2022.xlsx",
    2023: "contratos2023.xlsx",
}
# Criar DataFrames para cada ano
dfs = {}
for ano, arquivo in arquivos.items():
    dfs[ano] = pd.read_excel(data_dir + arquivo)

df_2021 = dfs[2021]
df_2022 = dfs[2022]
df_2023 = dfs[2023]
```

```python
arquivos = {
    2024: "contratos2024.xlsx",
    2025: "contratos2025.xlsx",
    2026: "contratos2024.xlsx",
}
# Criar DataFrames para cada ano
dfs = {}
for ano, arquivo in arquivos.items():
    dfs[ano] = pd.read_excel(data_dir + arquivo)

df_2024 = dfs[2024]
df_2025 = dfs[2025]
df_2026 = dfs[2026]
```

```python
df_total = pd.concat([df_2015,df_2016,df_2017,df_2018,df_2019,df_2020,df_2021,df_2022,df_2023,df_2024,df_2025,df_2026], ignore_index=True)

df_total.to_csv("contratos_total.csv", index=False, encoding="utf-8-sig")
```
